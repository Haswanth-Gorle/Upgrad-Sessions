# .iloc in Pandas

Purely integer-location based indexing for selection by position.

.iloc[] is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.

The end point will never be inclusive whereas in .loc the end point was inclusive


```python
df
```

```python
df.loc[1:4,['Name','Credit_Score']]
```

```python
df.iloc[1:4,0:2]
```

# Groupby in Pandas

```python
data={"Company":['GOOGLE','GOOGLE',"MSFT","MSFT","FB","FB","Amazon","Amazon"],
      "Employee_Name":['Neema',"Vrishank",'Sai','Jastinder','Vasvi','Sibasish','Ravi',"Anshuman"],
      "Sales":[200,120,300,400,500,130,230,155]}
df=pd.DataFrame(data)
df
```

```python
df.groupby('Company').size()
```

```python
pd.DataFrame(df.groupby('Company').sum()).sort_values(by='Sales',ascending=False)
```

```python
df
```

```python
df[['Company','Sales']].groupby('Company').max()
```

```python
#Sales between 300 and 600
df[ (df['Sales']>=300) & (df['Sales']<600)]
```

```python
df
```

```python
list_sales=list(df['Sales'])
list_sales.sort(reverse=True)
highest_sales=list_sales[1]
```

```python
df[df['Sales']==highest_sales]
```

# Merge in Pandas
