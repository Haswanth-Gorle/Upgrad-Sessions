.iloc[] is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.

In .iloc the endpoint is not inclusive

We can select the columns based on integer index

df.iloc[integer index of row,integer index of column]

# Groupby Function in Pandas

```python
data=pd.DataFrame({"Companies":['TCS','TCS','GOOGLE','MSFT','MSFT','GOOGLE'],
                  "Employee_name":['Arjun',"inder","Deepansh","Krishna",'John','Aditya'],
                   "Income":[1000,1500,2000,1500,1200,1500]})
data
```

```python
ipl=pd.DataFrame({"Teams":['MI','MI','RCB','RCB','CSK','CSK'],
                  "Player_Name":['Rohit',"Bumrah","Virat","Du_plesis",'Dhoni','Jadeha'],
                   "Income":[1000,1500,2000,1500,1200,1500]})
ipl
```

```python
ipl.groupby('Teams').size()
```

```python
ipl[['Teams','Income','Player_Name']].groupby('Teams')['Income','Player_Name'].max()
```

```python
ipl[['Teams','Income','Player_Name']].groupby('Teams')['Income','Player_Name'].min()
```

```python
data
```

```python
data.groupby('Companies').mean()
```

```python
data.groupby('Companies').sum()
```

```python
data.groupby('Companies').min()
```

```python
data.groupby('Companies').agg({"Income":['max','min','sum','std']})
```

# Merge in Pandas Dataframe

This will combine two or more dataframe based on a common column

```python
data
data['Income']=[1000,1100,1200,1300,1500,1600]
data
```

```python
df2={"Income":[1000,1100,1200,1300,1500,1600],"Employee_id":["tcs1",'tcs2','goog1','msft1','msft2','googl2']}
df2=pd.DataFrame(df2)
df2
```

```python
pd.merge(data,df2,on='Income')
```

```python
pd.concat([data,df2],axis=1)
```
