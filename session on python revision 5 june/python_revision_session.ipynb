```python
# Create a atm machine where the input of a pin should be a number and if 
#it is a number then withdraw the money and if the pin is incorrect 3 times then
#block the card

account_type='savings'
saving_account=10000
chance=3
index=1
real_pin='1347'
print("Welcome to the atm")

while(index<=chance):
    bank_account=input("Please choose the type of bank account")
    if bank_account==account_type:
        withdrawal_amt=input("Please input the withdrawal amount")
        if withdrawal_amt.isnumeric():
            index+=1
            withdrawal_amt=int(withdrawal_amt)
            if saving_account-withdrawal_amt>=0:
                pin_no=input("please enter your pin")
                if pin_no.isnumeric():
                    if pin_no==real_pin:
                        saving_account=saving_account-withdrawal_amt
                        print("Successful withdrawal of {}".format(saving_account))
                        break
                    else:
                        print("You have entered incorrect pin please try again")
                else:
                    continue
            else:
                print("You do not have suffient balance please add the correct amount")
                continue
print("Thank you for using the atm")

                    
        
```

```python
#Create a calculator function with three paramater num1,num2,operator=+,-,*,/

def calculator(num1,num2,operator):
    if operator=='+':
        Result=num1+num2
        return "the answer is "+str(Result)
    if operator=='-':
        Result=num1-num2
        return "the answer is "+str(Result)
    if operator=='*':
        Result=num1*num2
        return "the answer is "+str(Result)
    if operator=='/':
        try:  
            Result=(num1/num2)
            return "the answer is "+str(Result)
        except Exception as e:
            print("the error is--->" ,e)
        finally:
            print("Executed division")

while(True):
    try:
        num1=int(input("Please enter the first number")) #input by default is a string
        num2=int(input("Please enter the second number"))
    except Exception as e:
        print("error",str(e))
        continue
    operator=input("Please input the operator for calculator")
    if operator=='+' or operator=='-' or operator=='*' or operator:
        print(calculator(num1,num2,operator))
        break
```
